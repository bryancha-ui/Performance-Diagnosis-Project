```python
B_piv = pd.pivot_table(B, index=['주문시간'], values = '주문건수', columns = '상품명')
B_piv=B_piv = B_piv[B_piv.count(axis=1) != 1]
B_piv=B_piv= B_piv[B_piv.sum(axis=1) !=0]
B_piv[:] = np.where(B_piv.notnull(), B_piv.columns, B_piv)

C_piv = pd.pivot_table(C, index=['주문시간'], values = '주문건수', columns = '상품명')
C_piv=C_piv = C_piv[C_piv.count(axis=1) != 1]
C_piv=C_piv= C_piv[C_piv.sum(axis=1) !=0]
C_piv[:] = np.where(C_piv.notnull(), C_piv.columns, C_piv)

D_piv = pd.pivot_table(D, index=['주문시간'], values = '주문건수', columns = '상품명')
D_piv=D_piv = D_piv[D_piv.count(axis=1) != 1]
D_piv=D_piv= D_piv[D_piv.sum(axis=1) !=0]
D_piv[:] = np.where(D_piv.notnull(), D_piv.columns, D_piv)

E_piv = pd.pivot_table(E, index=['주문시간'], values = '주문건수', columns = '상품명')
E_piv=E_piv = E_piv[E_piv.count(axis=1) != 1]
E_piv=E_piv= E_piv[E_piv.sum(axis=1) !=0]
E_piv[:] = np.where(E_piv.notnull(), E_piv.columns, E_piv)

F_piv = pd.pivot_table(F, index=['주문시간'], values = '주문건수', columns = '상품명')
F_piv=F_piv = F_piv[F_piv.count(axis=1) != 1]
F_piv=F_piv= F_piv[F_piv.sum(axis=1) !=0]
F_piv[:] = np.where(F_piv.notnull(), F_piv.columns, F_piv)
```

```python
def f(pi):
    df_list= pi.values.tolist()
    Lis= []
    for i in range(0,len(pi)):
        li=df_list[i]
        cleanedli= [x for x in li if str(x) != 'nan']
        a=', '.join(cleanedli)
        Lis.append(a)
    pi['Comb']=Lis 
```

```python
f(B_piv)
f(C_piv)
f(D_piv)
f(E_piv)
f(F_piv)
```

```python
GNcount= B_piv['Comb'].value_counts().to_frame()
GNcount= GNcount.rename(columns={'Comb':'GNcount'})
GNsupp= B_piv['Comb'].value_counts(normalize=True).to_frame()
GNsupp= GNsupp.rename(columns={'Comb':'GNsupport'})

GGcount= C_piv['Comb'].value_counts().to_frame()
GGcount= GGcount.rename(columns={'Comb':'GGcount'})
GGsupp= C_piv['Comb'].value_counts(normalize=True).to_frame()
GGsupp= GGsupp.rename(columns={'Comb':'GGsupport'})

STcount= D_piv['Comb'].value_counts().to_frame()
STcount= STcount.rename(columns={'Comb':'STcount'})
STsupp= D_piv['Comb'].value_counts(normalize=True).to_frame()
STsupp= STsupp.rename(columns={'Comb':'STsupport'})

UJcount= E_piv['Comb'].value_counts().to_frame()
UJcount= UJcount.rename(columns={'Comb':'UJcount'})
UJsupp= E_piv['Comb'].value_counts(normalize=True).to_frame()
UJsupp= UJsupp.rename(columns={'Comb':'UJsupport'})

SCCcount= F_piv['Comb'].value_counts().to_frame()
SCCcount= SCCcount.rename(columns={'Comb':'SCCcount'})
SCCsupp= F_piv['Comb'].value_counts(normalize=True).to_frame()
SCCsupp= SCCsupp.rename(columns={'Comb':'SCCsupport'})
```
