```python
###Confidenced 구하는 함수
def conf(x,y,z):
    con=[]
    lif=[]
    for i in range (10):
        if x.index[i].count(',') == 1:
            cf= x.iloc[:, 0][i]/((y["상품명"].loc[y.index ==x.index[i].split(',')[0]]))[0]
            lf= cf/x.iloc[:, 1][i]
            con.append(cf)
            lif.append(lf)
        else: 
            ca=x.iloc[:, 0][i]/x.iloc[:, 0].loc[x.index == (x.index[i].split(',')[0] +","+ x.index[i].split(',')[1])][0]
            con.append(ca)
            la= ca/ x.iloc[:, 1][i]
            lif.append(la)
    z['confidence']= con
    z['lift']= lif
    
            
            
            
```
